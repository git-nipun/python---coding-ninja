```python
class Student:
    def __init__(self,name,rollNumber):
        self.name = name
        self.rollNumber = rollNumber
        
    def printStudent(self):
        print('My name is ', self.name, 'and my roll number is ',self.rollNumber)
```

```python
s1 = Student('Ankush',12)
```

```python
s1.__dict__
```

```python
s1.printStudent()
```

```python
Student.printStudent(s1)
```
