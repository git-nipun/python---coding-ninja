split

```python
str = "My name is Parikh"
li = str.split()
print(li)
```

```python
str = "My,name,is,Parikh"
li = str.split()
print(li)
```

```python
str = "My,name,is,Parikh"
li = str.split(',')
print(li)
```

```python
str = "My,name,is,Parikh"
li = str.split(',',1)
print(li)
```

```python
str = "My,name,is,Parikh"
li = str.split(',',2)
print(li)
```

REplace

```python
str = "My name is Parikh"
str = str.replace("Parikh","Rohan")
print(str)
```

```python
str = "My name is Parikh"
str = str.replace("which","Rohan")
print(str)
```

```python
str = "My name is Parikh Parikh Parikh"
str = str.replace("Parikh","Rohan")
print(str)
```

```python
str = "My name is Parikh Parikh Parikh"
str = str.replace("Parikh","Rohan",2)
print(str)
```
