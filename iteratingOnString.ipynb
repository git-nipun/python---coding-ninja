how to itearate on string

```python
str = "Hello World"
```

```python
count = 0
for letter in str:
    if (letter == 'l'):
        count = count + 1
print(count)
```

```python
count = 0
for i in range(len(str)):
    if(str[i] == 'l'):
        count = count + 1
print(count)
```
