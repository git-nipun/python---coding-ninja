```python
n = int(input())

for i in range(1,n+1,1):
    
    # spaces
    for spaces in range(0,n-i,1):
        print('-',end='')
    # increasing 
    for inc in range(i,2*i,1):
        print(inc,end='')
    # decreasing
    for dec in range(2*i-2,i-1,-1):
        print(dec,end='')
        
    print()
```

```python
# break statement , chk prime, while loop
n = int(input())

div = 2
flag = False
while(div<n):
    if n%div==0:
        flag = True
        break
    div = div + 1

if flag:
    print('not prime')
else:
    print('prime')
```

```python
# break statement, chk prime, For loop

n = int(input())
flag = False
for div in range(2,n,1):
    if n%div==0:
        flag = True
        break
        

if flag:
    print('not prime')
else:
    print('prime')
```

```python
# print Prime from 2 to n

n = int(input())
k = 2
while(k<=n):
    d=2
    flag = False
    while(d<k):
        if k%d==0:
            flag = True
            break
        d = d + 1
    if (not(flag)):
        print(k)
    k = k + 1
```

```python
# print Prime Numbers from 2 to N

n = int(input())
num = 2
while(num<=n):
    div = 2
    flag = False
    while(div<num):
        if(num%div==0):
            flag = True
            break
        div = div + 1
            
    if(not(flag)):
        print(num)
    num = num + 1
```
