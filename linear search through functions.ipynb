linear search through functions

```python
def linear_search(li,ele):
    # li is the list and ele is the element
    for i in range(len(li)):
        if li[i] == ele:
            return i
    return -1

li = [1,2,3,6,5]
index = linear_search(li,9)
print(index)
```

# reverse a list

```python
def reverse_l(li):
    length = len(li)
    for i in range(length//2):
        li[i],li[length-1-i] = li[length-1-i],li[i]

li = [1,2,3,4,5,6,7]
reverse_l(li)
print(li)
```

```python
def reverse_l_2(li):
    length = len(li)
    for i in range(length//2):
        li[i],li[-i-1] = li[-i-1],li[i]

li = [1,2,3,4,5,6,7]
reverse_l_2(li)
print(li)
```
