# linear search through functions

```python
def linear_search(li,ele):
    for i in range(len(li)):
        if ele == li[i]:
            return i
    
    return -1

li = [1,2,3,6,7,8]
index = linear_search(li,6)
print(index)
```

# reverse a list

```python
def reverse_list(li):
    length = len(li)
    for i in range(length//2):
        li[i],li[length-i-1] = li[length-i-1],li[i]

li = [1,2,3,4,5,6]
reverse_list(li)
print(li)
```

```python
# using negative indexing 
def reverse_list(li):
    length = len(li)
    for i in range(length//2):
        li[i],li[-i-1] = li[-i-1],li[i]

li = [1,2,3,4,5,6]
reverse_list(li)
print(li)
```

 Swap Alternate
Send Feedback
You have been given an array/list(ARR) of size N. You need to swap every pair of alternate elements in the array/list.
You don't need to print or return anything, just change in the input array itself.
Input Format :

The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.

First line of each test case or query contains an integer 'N' representing the size of the array/list.

Second line contains 'N' single space separated integers representing the elements in the array/list.

Output Format :

For each test case, print the elements of the resulting array in a single row separated by a single space.

Output for every test case will be printed in a separate line.

Constraints :

1 <= t <= 10^2
0 <= N <= 10^5
Time Limit: 1sec

Sample Input 1:

1
6
9 3 6 12 4 32

Sample Output 1 :

3 9 12 6 32 4

Sample Input 2:

2
9
9 3 6 12 4 32 5 11 19
4
1 2 3 4

Sample Output 2 :

3 9 12 6 32 4 11 5 19 
2 1 4 3 



```python
def swap_alternate(n,li):
#     length = n
    for i in range(0,n-1,2):
        li[i],li[i+1] = li[i+1],li[i]
```

```python
def print_list(n,li):
    for i in range(n):
        print(li[i], end=" ")
    print()
```

```python
def take_input():
    n = int(input())
    
    if n == 0:
        return list(),0
    
    li = [int(x) for x in input().split()]
    return n,li
```

```python
t = int(input())
while(t>0):
    n,li = take_input()
    if(n!=0):
        swap_alternate(n,li)
        print_list(n,li)
    t = t - 1
```

# find unique

```python
def find_unique(li,n):
    for i in range(n):
        j = 0
        while(j < n):
            if i != j:
                if li[i] == li[j]:
                    break
            j = j + 1
        if j == n:
            return li[i]
n = 7
li = [2,3,1,6,3,6,2]
ele = find_unique(li,n)
print(ele)
```

# find duplicate

```python
def find_duplicate(n,li):
    for i in range(n-1):
        for j in range(i+1,n):
            if li[i] == li[j]:
                return li[i]
```

```python
n = int(input())
li = [int(x) for x in input().split()]
print(find_duplicate(n,li))
```

# pair sum

```python
def pair_sum(size,arr,sum):
    num_pair = 0
    for i in range(size-1):
        for j in range(i+1,size):
            if(arr[i]+arr[j]==sum):
                num_pair = num_pair + 1
    return num_pair
#                 print(arr[i]," ",arr[j])
```

```python
size = int(input())
arr = [int(x) for x in input().split()]
sum = int(input())
print(pair_sum(size,arr,sum))
```

# triplet sum

```python
def triplet_sum(size,arr,sum):
    num_triplets = 0
    for i in range(size):
        for j in range(i+1,size):
            for z in range(j+1,size):
                
                if(arr[i]+arr[j]+arr[z] == sum):
                    num_triplets = num_triplets + 1
    return num_triplets
#                     print(arr[i]," ",arr[j], " ",arr[z])
```

```python
size = int(input())
arr = [int(x) for x in input().split()]
sum = int(input())
print(triplet_sum(size,arr,sum))
```
