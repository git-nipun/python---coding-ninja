# print all words with frequency K

```python
s = "this is a word string having many many word"
k = 2
```

```python
words =s.split()
```

```python
words
```

```python
# method 1
d = {}
for w in words:
    if w in d:
        d[w] = d[w] + 1
    else:
        d[w] = 1
```

```python
d
```

```python
# method 2
d = {}
for w in words:
    d[w] = d.get(w,0) + 1
```

```python
d
```

```python
for w in d:
    if d[w] == k:
        print(w)
```

```python
def print_k_frequency_words(s,k):
    words = s.split()
    d = {}
    for w in words:
        d[w] = d.get(w,0) + 1
    for w in d:
        if d[w] == k:
            print(w)

print_k_frequency_words(s,k)
```

```python
def sum_unique(l):
    s = set()
    for i in l:
        s.add(i)
    sum = 0
    for i in s:
        sum = sum + i
    return sum
```

```python
sum_unique([1,2,3,4,2,3,5,6,3,4,3,4,3,4])
```
