```python
n = int(input())

i = 1 
while(i <= n):
    j = 1
    while(j <= n):
        charP = chr(ord('A') + j - 1)
        print(charP, end='')
        j = j + 1
    print()
    i = i + 1
```

```python
n = int(input())

i = 1 
while(i <= n):
    j = 1
    star_char = chr(ord('A') + i - 1)
    while(j <= n):
        charP = chr(ord(star_char) + j - 1)
        print(charP, end='')
        j = j + 1
    print()
    i = i + 1
```
