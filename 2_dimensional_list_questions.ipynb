# row wise sum

```python
# from sys import stdin
def row_wise_sum(n,m,mat):
    for i in range(n):
        row_sum = 0
        for j in range(m):
            row_sum += mat[i][j]
        print(row_sum, end=' ')
    

str = input().split()
n,m = int(str[0]),int(str[1])
mat = [[int(j) for j in input().split()]for i in range(n)]
row_wise_sum(n,m,mat)
```

# largest row or column

```python
min_value = -2**31
def lar_row_col(n_rows,m_colm,mat):
    is_row = True
    largest_sum = min_value
    num = 0
    
    for i in range(n_rows):
        row_sum = 0
        for j in range(m_colm):
            row_sum += mat[i][j]
        if row_sum > largest_sum:
            largest_sum = row_sum
            num = i
    
    for j in range(m_colm):
        col_sum = 0
        for i in range(n_rows):
            col_sum += mat[i][j]
        if col_sum > largest_sum:
            largest_sum = col_sum
            num = j
            is_row = False

    if is_row:
        print('row' , num , largest_sum)
    else:
        print('col' , num , largest_sum)
    

str = input().split()
n,m = int(str[0]),int(str[1])
mat = [[int(j) for j in input().split()]for i in range(n)]
lar_row_col(n,m,mat)
```

# wave print

```python
def wave_print(nRows,mCols,mat):
    for j in  range(mCols):
        for i in range(nRows):
            if j % 2 == 0:
                print(mat[i][j])
            else:
                print(mat[i][-1])
    
str = input().split()
n,m = int(str[0]),int(str[1])
mat = [[int(j) for j in input().split()]for i in range(n)]
wave_print(n,m,mat)
```
