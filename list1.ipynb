```python
li = [1,2,3,4,5,6,7,8,9]
li_new = []
for ele in li:
    if ele%2==0 and ele%3==0:
        li_new.append(ele)
    
print(li_new)
```

```python
li_new_2 = [ele for ele in li if ele%2==0 and ele%3==0]
```
