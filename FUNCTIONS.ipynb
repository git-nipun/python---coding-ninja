```python
def fact(a):
    a_fact = 1
    for i in range(1,a+1):
        a_fact = a_fact * i
    return a_fact
```

```python
fact(3)
```

```python
n = int(input())
r = int(input())

n_fact = fact(n)
r_fact = fact(r)
n_r_fact = fact(n-r)

ans = n_fact//(r_fact * n_r_fact)
print(ans)
```

```python
n = int(input())
r = int(input())

n_fact = 1
for i in range(1,n+1):
    n_fact = n_fact * i
    
r_fact = 1
for i in range(1,r+1):
    r_fact = r_fact * i
    
n_r_fact = 1
for i in range(1,n-r+1):
    n_r_fact = n_r_fact * i
    
ans = n_fact//(r_fact * n_r_fact)
print(ans)
```

```python
def isPrime(n):
    for d in range(2,n):
        if(n % d == 0):
            break
        else:
            return True
        return False
```

```python
isPrime(17)
```

```python
def primeFrom2toN(n):
    for k in range(2,n+1):
        # check if k is peime and if it is print k
        is_k_prime = isPrime(k)
        if(is_k_prime):
            print(k)
```

```python
primeFrom2toN(20)
```

```python
def ncr(n , r):
    n_fact = fact(n)
    r_fact = fact(r)
    n_r_fact = fact(n-r)

    ans = n_fact//(r_fact * n_r_fact)
    return ans
```
