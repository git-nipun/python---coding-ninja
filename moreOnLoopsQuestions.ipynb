# binary Pattern

```python
n = int(input())

for i in range(1,n+1,1):
    for j in range(1,n-i+2,1):
        if i % 2 == 0: 
            print('0',end = '') 
        else : 
            print('1',end = '') 
    print()
```

# print number pyramid

```python
# print number pyramid WHILE LOOP

n = int(input())

i = 1
while(i<=n):
    # spaces
    spaces = 1
    while(spaces<=i-1):
        print('-',end='')
        spaces = spaces+1
        
    #numbers
    j = 1
    num = i
    while(j<=n-i+1):
        print(num,end='')
        j = j + 1
        num = num + 1
        
    print()
    i = i + 1
    
i = 1
while(i<n):
    #spaces 
    spaces = 1
    while(spaces<n-i):
        print("-",end='')
        spaces = spaces + 1
    #numbers
    j = 1
    num = n-i
    while(j<=i+1):
        print(num, end='')
        j = j + 1 
        num = num + 1
    
    print()
    i = i + 1
```

```python
# print number pyramid FOR LOOP

n = int(input())

for i in range(1,n+1,1):
    #spaces
    for spaces in range(1,i-1+1,1):
        print("-", end='')
        
    # numbers    
    for num in range(i,n+1,1):
        print(num, end="")
    
    print()
    
for i in range(1,n,1):
    #spaces
    for spaces in range(1,n-i,1):
        print("-",end='')
        
    #numbers
    for num in range(n-i,n+1,1):
        print(num, end="")
        
    print()
    
```

# diamond of stars

```python
n = int(input())
n1 = n//2+1
for i in range(1,n1+1,1):
    #spaces
    for spaces in range(1,n//2+2-i,1):
        print("-",end="")
    #stars
    for star in range(1,2*i,1):
        print("*",end="")
    
    print()

n2 = n//2
for i in range(n2,0,-1):
    #spaces
    for spaces in range(1,n1-i+1,1):
        print("-",end="")
        
    #stars
    for stars in range(1,2*i,1):
        print("*", end='')
        
    print()
```
