```python
li = [1,2,3,4,5,6,7,8,9]
li_new = []
for ele in li:
    if ele%2==0:
        if ele%3==0:
            li_new.append(ele)
print(li_new)
```

```python
li_1 = [1,2,3,4,5]
li_2 = [2,4,6,7]
li_inter = []
for ele in li_1:
    for ele_2 in li_2:
        if ele == ele_2:
            li_inter.append(ele)
print(li_inter)
```

```python
li_1 = [1,2,3,4,5]
li_2 = [2,4,6,7]
li_inter = [ele for ele in li_1 for ele2 in li_2 if ele==ele2]
```

```python
print(li_inter)
```
