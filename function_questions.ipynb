```python
# check prime
def isPrime(n):
    
    for div in range(2,n,1):
        if(n%div == 0):
            break
    else:
        return True
        
    return False
            
    
```

```python
isPrime(17)
```

```python
# print all prime numbers from 2 to n

def prime_2_to_n(n):
    for k in range(2,n+1):
        is_k_prime = isPrime(k)
        
        if(is_k_prime):
            print(k)
```

```python
prime_2_to_n(20)
```
