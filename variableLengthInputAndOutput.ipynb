```python
def sum2(a,b,*more):
    print(a)
    print(b)
    print(more)
    print(type(more))#empty tuple
    ans = a+b
    for i in more:
        ans = ans + i
    return ans
```

```python
sum2(2,3,4,5)
```

```python
def sum_diff(a,b):
    return a+b,a-b,a*b
```

```python
d,e = sum_diff(4,1)
print(d)
print(e)
```

```python
d,e,f = sum_diff(4,1)
print(d)
print(e)
print(f)
```
