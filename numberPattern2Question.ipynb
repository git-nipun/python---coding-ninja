```python
n = int(input())

# upper half
i = 1
n1 = n//2 + 1
while(i <= n1):
    #spaces 
    spaces = 1
    while(spaces<=i-1):
        print(" ", end="")
        spaces = spaces + 1
    # stars
    stars = 1
    while(stars<=i):
        print("*",end=" ")
        stars = stars+1

    print()
    i = i + 1

# lower half
i = 1
n2 = n//2
while(i<=n2):
    # spaces 
    spaces = 1
    while(spaces <= n2-i):
        print(" ",end="")
        spaces = spaces+1
    # stars
    stars = 1
    while(stars <= n2-i+1):
        print("*", end=" ")
        stars = stars + 1
    
    print()
    i = i + 1
```
