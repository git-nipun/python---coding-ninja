# binary search

```python
def binary_search(arr, element):
    start = 0
    end = len(arr) -1
    
    while(start<=end):
        mid = (start + end)//2
        
        if(arr[mid] == element):
            return mid
        elif(arr[mid] > element):
            end = mid - 1
        else:
            start = mid + 1
        
    return -1
    
```

```python
arr = [1,2,3,4,5,6,7]
# length = len(li)
element = 9
index = binary_search(arr, element)
print(index)
```

```python
def bubble_sort(arr):
    length = len(arr)
    for i in range(length-1):
        for j in range(length-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
def printList(arr, n) :
    for i in range(n) :
        print(arr[i], end = " ") 
    print() 
        
arr = [6,4,8,3,1,2]
bubble_sort(arr)
print(arr)
# printList(arr)
# print(arr)
```

# merge two sorted array

```python
def merger_2_sorted_array(arr1,arr2):
    new_arr = []
    i = 0
    j = 0
    while(i < len(arr1) and j < len(arr2)):
        if(arr1[i]<arr2[j]):
            new_arr.append(arr1[i])
            i = i + 1
        elif(arr1[i]==arr2[j]):
            new_arr.append(arr1[i])
            new_arr.append(arr2[j])
            i = i + 1
            j = j + 1
        else:
            new_arr.append(arr2[j])
            j = j + 1
    while(i<len(arr1)):
        new_arr.append(arr1[i])
        i = i + 1
    while(j<len(arr2)):
        new_arr.append(arr2[j])
        j = j + 1
    return new_arr
            

arr1 = [10,100,500]
# length_1 = len(arr1)
arr2 = [4,7,9,25,30,300,450]
# length_2 = len(arr2)
new_arr = merger_2_sorted_array(arr1,arr2)
print(new_arr)
```

```python
def reverse_array(arr,n):
    for i in range(n//2):
        arr[i],arr[-i-1] = arr[-i-1],arr[i]

def rotate_array(arr,n,d):
    for i in range((n-d)//2):
        arr[i],arr[n-i-d-1] = arr[n-i-d-1],arr[i]
    for i in range(d//2):
        arr[-i-1],arr[-i-d] = arr[-i-d],arr[-i-1]
    
n = 7
arr = [1,2,3,4,5,6,7]
d = 2
reverse_array(arr,n)
print(arr)
rotate_array(arr,n,d)
print(arr)
```

# second largest in array

```python
def second_largest_inArr(arr,n):
    largest = -2**31
    second_largest = -2**31
    for i in arr:
        if i == largest:
            i = i + 1
            continue
        elif i > largest or i > second_largest:
            second_largest = largest
            largest = i
    return second_largest
        

n = 7
arr = [2,6,1,9,8,9,7]
num = second_largest_inArr(arr,n)
print(num)
```

```python
MIN_VALUE = -2147483648 
def secondLargestElement(arr, n):
    if n == 0 : 
        return MIN_VALUE
    largest = arr[0] 
    secondLargest = MIN_VALUE 
    for i in range(n) : 
        if largest < arr[i] : 
            secondLargest = largest 
            largest = arr[i]
        elif secondLargest < arr[i] and arr[i] != largest :
            secondLargest = arr[i] 
    return secondLargest
```

# check array rotation

```python
def check_arr_rotation(arr,n):
    for i in range(n-1):
        if arr[i] > arr[i+1]:
            return i + 1
    return 0

n = 6
arr = [5,6,1,2,3,4]
index = check_arr_rotation(arr,n)
print(index)
```
