```python
def selectionSort(arr):
    length = len(arr)
    for i in range (length-1):
        minIndex = i
        for j in range(i+1,length):
            if(arr[j] < arr[minIndex]):
                minIndex = j
        arr[i],arr[minIndex] = arr[minIndex],arr[i]
            
n = int(input())
# arr = [1,3,2,4,0,6,8]
arr = [int(x) for x in input().split()[:n]] 
selectionSort(arr)
# print(arr)
for ele in arr:
    print(ele,end=' ')
```
