```python
def chk_permutaion(str1,str2):
    if len(str1) != len(str2):
        return False
    frequency = [0]*256
    for i in range(len(str1)):
        ch = ord(str1[i])
        frequency[ch]+=1
    
    for i in range(len(str2)):
        ch = ord(str2[i])
        frequency[ch]-=1
        
    for i in range(256):
        if frequency[i] != 0:
            return False
    
    return True

str1 = "aba"
str2 = "ba"
ans = chk_permutaion(str1,str2)
if(ans):
    print("true")
else:
    print("false")
```

# Remove Consecutive Duplicates 

```python
def remove_consecutive_dupli(str):
    new_str = ""
    for i in str:
        for j in (i+1,str):
            if i == j:
                new_str = new_str + i
            else:
                new_str = new_str + i
    return new_str

str = "aabccbaa"
print(remove_consecutive_dupli(str))
```

```python
def remove_consecutive_dupli(string):
    n = len(string)
    if n == 0:
        return string
    
    answer = ""
    
    start_index = 0
    while start_index < n:
        unique_char = string[start_index]
        next_unique_char_index = start_index + 1
        
        while (next_unique_char_index < n) and (string[next_unique_char_index] == unique_char):
            next_unique_char_index += 1
            
        answer += unique_char
        start_index = next_unique_char_index
    return answer

string = "aabccbaa"
ans = remove_consecutive_dupli(string)
print(ans)
```

# reverse each word

```python
def reverse_word(string, start, end):
    reverse = ""
    while start<end:
        reverse = string[start] + reverse
        start = start + 1
    
    return reverse

def reverse_each_word(string):
    n = len(string)
    previous_space_index = -1
    
    ans = ""
    
    i = 0
    while(i<n):
        if string[i] == " ":
            ans = ans + ( reverse_word ( string , previous_space_index + 1 , i ) + " ")
            previous_space_index = i
        i = i + 1
    ans = ans + (reverse_word ( string , previous_space_index + 1 , i ) + " ")
    
    return ans

string = "Welcome to Coding Ninjas"
ans = reverse_each_word(string)
print(ans)
```

# Remove character

```python
def remove_all_occurence_of_char(string,ch):
    n = len(string)
    new_str = ''
    for i in range(n):
        if(string[i] != ch):
            new_str += string[i]
    return new_str

string = "xxyyzxx"
char = 'y'
ans = remove_all_occurence_of_char(string,char)
print(ans)
```

# Highest Occuring Character 

```python
def highest_Occuring_Char(string):
    n = len(string)
    frequency = [0]*256
    max_frequency = 0
    
    for i in range(n):
        frequency[ord(string[i])] += 1
        max_frequency = max(max_frequency ,frequency[ord(string[i])])
    
    answer = '\0'
    
    for i in range(n):
        frequency[ord(string[i])] == max_frequency
        answer = string[i]
        break
    return answer

string = "abdefgbabfba"
ans = highest_Occuring_Char(string)
print(ans)
```

# Compress the String

```python
def compress_string(string):
    n = len(string)
    answer = ""
    
    if n == 0:
        return answer
    current_char_count = 1
    answer += string[0]
    
    for i in range(1,n):
        if string[i] == string[i-1]:
            current_char_count += 1
        
        else:
            if current_char_count > 1:
                answer += str(current_char_count)
                current_char_count = 1
            answer += string[i]
            
    if current_char_count > 1:
        answer += str(current_char_count)
            
    return answer

string = "aaabbccdsa"
# o/p = a3b2c2dsa
ans = compress_string(string)
print(ans)
```
