### print SUM of all EVEN numbers

```python
# print sum of EVEN n natural numbers

n = int(input())
count = 1
sum = 0

while count<=n:
    if(count % 2 == 0):
        sum = sum + count
    count = count+1
print(sum)
```

### fahrenheit to celcius

### reverse of a number

```python
n = int(input())
rev = 0 
while(n!=0):
    last = n%10
    n = n // 10
    rev = rev*10 + last

print(rev)
```

### palindrome number

```python
num = int(input())
temp = num
rev = 0 
while(temp!=0):
    rev = rev*10 + temp%10
    temp = temp // 10

if(num == rev):
    print('true')
else:
    print("false")
```

### sum of even and odd digits of a number

```python
num = int(input())
temp = num

Sum_Even = 0
Sum_Odd = 0

while(temp!=0):
    digit = temp%10
    if(digit%2==0):
        Sum_Even = Sum_Even + digit
        
    else:
        Sum_Odd = Sum_Odd + digit
        
    temp = temp // 10
    
    
print(Sum_Even, " ", Sum_Odd)
    
```
