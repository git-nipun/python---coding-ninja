## Fraction class 3

```python
class Fraction:
    def __init__(self, num = 0, den = 1):
        if den == 0: 
            #Throw error
            den = 1
        self.num = num
        self.den = den
        
    def print(self):
        if self.num == 0:
            print(0)
        elif self.den == 1:
            print(self.num)
        else:
            print(self.num, '/', self.den)
            
    def simplify(self):
        if self.num == 0:
            self.den == 1
            return
        
        current = min(self.num, self.den)
        while current > 1:
            if self.num % current == 0 and self.den % current == 0:
                break
            current -= 1
        self.num = self.num//current
        self.den = self.den//current
    
    def add(self, otherFraction):
        newNumerator = otherFraction.den*self.num + otherFraction.num*self.den    # bx + ay
        newDenominator = otherFraction.den * self.den                             # b*y 
        
        self.num = newNumerator
        self.den = newDenominator
        self.simplify()
        
    def multiply(self, otherFraction):
        newNumerator = otherFraction.num * self.num 
        newDenominator = otherFraction.den * self.den
        self.num = newNumerator
        self.den = newDenominator
        self.simplify()
        
        
```

```python
f1 = Fraction(2,3)
f2 = Fraction(1,5)
```

```python
f1.multiply(f2)
f1.print()
```

```python
f1 = Fraction(7,3)
f2 = Fraction(1)
```

## Complex Number Addition, Multiplication and Conjucntion

```python
class cNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
    
    def print(self):
        if self.imag < 0:
            print(self.real, '- i', end = "")
            print(self.imag* -1)
        else:
            print(self.real, '+ i', end = "")
            print(self.imag)
              
    def add(self, otherNumber):
        newReal = self.real + otherNumber.real
        newImag = self.imag + otherNumber.imag
        self.real = newReal
        self.imag = newImag
    
    def multiply(self, otherNumber):
        newReal = self.real * otherNumber.real - self.imag * otherNumber.imag
        newImag = self.imag * otherNumber.real + self.real * otherNumber.imag
        self.real = newReal
        self.imag = newImag
    
    def conjunction(self):        # Conjunction returns 'a - ib' for the input 'a + ib'
        self.real = self.real
        self.imag = self.imag * -1
    
```

```python
c1 = cNumber(4,5)
c2 = cNumber(1,2)
```

```python
c1.add(c2)
c1.print()
```

```python
c1.multiply(c2)
c1.print()
```

```python
c3 = cNumber(2,3)
c3.conjunction()
c3.print()
```
